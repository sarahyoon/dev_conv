```python
import pandas as pd
import numpy as np

# ==============================
# 1. 설정
# ==============================

# 검사할 파일 경로
file_path = "900.csv"   # ← 여기만 바꿔서 사용

# 필수 컬럼 목록 (현재 데이터 구조 기준)
REQUIRED_COLS = [
    "Date",
    "CustomerID",      # 없으면 나중에 추가해도 됨
    "MaskedSub",
    "MeterCategory",
    "MeterSubCategory",
    "MeterName",
    "ChargeType",
    "EffectivePrice",
    "Quantity",
    "TotalCost",
]

# ==============================
# 2. 로딩 및 기본 정보
# ==============================

print(f"===== 파일 로딩: {file_path} =====")
df = pd.read_csv(file_path)
print("행(row) 수:", len(df))
print("컬럼 목록:", list(df.columns))
print()

# ==============================
# 3. 필수 컬럼 체크
# ==============================

missing_cols = [c for c in REQUIRED_COLS if c not in df.columns]
if missing_cols:
    print("⚠ 필수 컬럼 누락:", missing_cols)
else:
    print("✅ 필수 컬럼 모두 존재")
print()

# ==============================
# 4. Date 컬럼 확인 및 변환
# ==============================

if "Date" in df.columns:
    # 문자열/정수 → datetime 변환 시도
    df["Date_raw"] = df["Date"]  # 원본 백업

    
    # 1. 원본 텍스트 확보
    raw = df["Date"].astype(str)
    
    # 2. 뒤 8자리 숫자(YYYYMMDD)만 추출
    yyyymmdd = raw.str.extract(r"(\d{8})$")[0]
    
    # 3. 날짜로 정확히 변환
    df["Date"] = pd.to_datetime(yyyymmdd, format="%Y%m%d", errors="coerce")
    
    # 4. 시간 제거 → 날짜만 유지
    df["Date"] = df["Date"].dt.date
    df["Date"] = pd.to_datetime(df["Date"])
    
    n_invalid = df["Date"].isna().sum()
    print("✅ Date 변환 완료")
    print(" - 전체 행:", len(df))
    print(" - 변환 실패(NaT) 행 수:", n_invalid)
    if n_invalid > 0:
        print("  ⚠ Date 형식이 이상한 행이 있습니다. (예: YYYYMMDD가 아닌 형식)")
    
    print(" - 날짜 범위:", df["Date"].min(), "~", df["Date"].max())
    print(" - 서로 다른 날짜 수:", df["Date"].nunique())
    print()
else:
    print("❌ Date 컬럼이 없음 - 이 파일은 시계열 분석에 바로 사용할 수 없습니다.")
    print()

# ==============================
# 5. 수치 컬럼 타입 & 기본 통계
# ==============================

for col in ["Quantity", "TotalCost", "EffectivePrice"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
        print(f"✅ {col} 컬럼 숫자 변환 완료")
        print(f" - NaN 개수: {df[col].isna().sum()}")
        print(f" - 최소값: {df[col].min()}, 최대값: {df[col].max()}")
        if (df[col] < 0).sum() > 0:
            print(f"  ⚠ {col}에 음수 값이 있습니다. (개수: {(df[col] < 0).sum()})")
        print()
    else:
        print(f"⚠ {col} 컬럼이 없습니다.")
        print()

# ==============================
# 6. ChargeType / MaskedSub / MeterCategory 분포
# ==============================

if "ChargeType" in df.columns:
    print("✅ ChargeType 분포:")
    print(df["ChargeType"].value_counts())
    print()

if "MaskedSub" in df.columns:
    print("✅ MaskedSub(구독) 개수:", df["MaskedSub"].nunique())
    print(df["MaskedSub"].value_counts().head())
    print()

if "MeterCategory" in df.columns:
    print("✅ MeterCategory 상위 10개:")
    print(df["MeterCategory"].value_counts().head(10))
    print()

# ==============================
# 7. Usage만 필터링한 후 기본 체크
# ==============================

if "ChargeType" in df.columns:
    df_usage = df[df["ChargeType"] == "Usage"].copy()
    print("✅ Usage 데이터만 필터링")
    print(" - Usage 행 수:", len(df_usage))
    print(" - 전체 대비 비율: {:.1f}%".format(len(df_usage) / len(df) * 100))
    print()
else:
    df_usage = df.copy()
    print("⚠ ChargeType이 없어 전체 데이터를 Usage처럼 취급합니다.")
    print()

# ==============================
# 8. RI(Reserved Instance) 의심 구간 체크
#    Cost == 0, Qty > 0, UnitPrice > 0
# ==============================

if all(c in df_usage.columns for c in ["TotalCost", "Quantity", "EffectivePrice"]):
    ri_mask = (
        (df_usage["TotalCost"] == 0) &
        (df_usage["Quantity"] > 0) &
        (df_usage["EffectivePrice"] > 0)
    )
    ri_count = ri_mask.sum()
    print("✅ RI 의심 구간 수:", ri_count)
    if ri_count > 0:
        print(" - 예시 5행:")
        print(df_usage[ri_mask].head())
    print()
else:
    print("⚠ RI 판단에 필요한 컬럼(Quantity, TotalCost, EffectivePrice)이 부족합니다.")
    print()

# ==============================
# 9. 일별 집계 테스트 (이 고객 데이터가 LSTM 입력으로 쓸 수 있는지 확인)
# ==============================

if all(c in df_usage.columns for c in ["Date", "MeterCategory", "TotalCost"]):
    daily = (
        df_usage
        .groupby(["Date", "MeterCategory"])
        .agg({
            "TotalCost": "sum",
            "Quantity": "sum",
            "EffectivePrice": "mean"
        })
        .reset_index()
    )
    print("✅ 일자 × 카테고리 집계 완료")
    print(" - 집계 후 행 수:", len(daily))
    print(" - 서로 다른 날짜 수:", daily["Date"].nunique())
    print(" - 예시 5행:")
    print(daily.head())
    print()
    
    # Pivot까지 한 번 해보기 (LSTM용 다변량 시계열 구조)
    pivot_cost = daily.pivot(
        index="Date",
        columns="MeterCategory",
        values="TotalCost"
    ).fillna(0)
    
    print("✅ Pivot 변환 완료 (LSTM 입력 후보)")
    print(" - shape (날짜 수, 카테고리 수):", pivot_cost.shape)
    print(" - 상위 5행:")
    print(pivot_cost.head())
    print()
else:
    print("❌ Date / MeterCategory / TotalCost가 부족해서 일별 집계를 만들 수 없습니다.")
    print()

print("===== 검증 체크리스트 종료 =====")
```

```python
print("✅ 핵심 컬럼 기준 중복 검사 (논리적 중복)")

key_cols = ["Date", "MaskedSub", "MeterName", "TotalCost", "Quantity"]

if all(col in df.columns for col in key_cols):
    dup_logic = df.duplicated(subset=key_cols)
    dup_logic_count = dup_logic.sum()

    print(" - 논리적 중복 행 수:", dup_logic_count)

    if dup_logic_count > 0:
        print("⚠ 논리적 중복 예시 (상위 5개):")
        print(df[dup_logic].head())
else:
    print("⚠ 중복 검사용 핵심 컬럼이 일부 없습니다:", key_cols)

print()
```
