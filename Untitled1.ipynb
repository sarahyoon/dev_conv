```python
import glob
import pandas as pd

files = glob.glob("*.csv")

dfs = []
for f in files:
    temp = pd.read_csv(f)
    dfs.append(temp)

df_all = pd.concat(dfs, ignore_index=True)
# 컬럼명이 데이터 행에 섞여 있는지 확인
print(df_all[df_all["Date"] == "Date"])
```

```python
import pandas as pd

df = pd.read_csv("total.csv")
```

```python
df.info()
df.head(10)
```

```python
# 1) CustomerID 컬럼 하나로 통일
if "customerId" in df_all.columns:
    df_all.rename(columns={"customerId": "CustomerID"}, inplace=True)

# 2) 분석에 필요 없는 컬럼 제거
df_all.drop(columns=["Date_raw", "sum_Cost"], errors="ignore", inplace=True)

# 3) 타입 정리
df_all["Date"] = pd.to_datetime(df_all["Date"])   # 이미 yyyy-mm-dd라면 안전하게 한 번 더
for col in ["TotalCost", "Quantity", "EffectivePrice"]:
    df_all[col] = pd.to_numeric(df_all[col], errors="coerce")

print(df_all.dtypes)
print(df_all.head())
```

```python
daily_cost = (
    df
    .groupby(["CustomerID", "Date"])["TotalCost"]
    .sum()
    .reset_index()
    .sort_values(["CustomerID", "Date"])
)

print(daily_cost.head())
print("고객 수:", daily_cost["CustomerID"].nunique())
```

```python
import matplotlib.pyplot as plt

sample = daily_cost[daily_cost["CustomerID"] == 900]

plt.figure(figsize=(12,5))
plt.plot(sample["Date"], sample["TotalCost"])
plt.title("Customer 100 Daily Cost Pattern")
plt.xlabel("Date")
plt.ylabel("Total Cost")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 고객 ID 목록 가져오기
customer_list = daily_cost["CustomerID"].unique()

# 상위 5명만 시각화
for cid in customer_list[:9]:
    data = daily_cost[daily_cost["CustomerID"] == cid]
    
    plt.figure(figsize=(12,4))
    plt.plot(data["Date"], data["TotalCost"])
    plt.title(f"Customer {int(cid)} Daily Cost Pattern")
    plt.xlabel("Date")
    plt.ylabel("Total Cost")
    plt.grid(True)
    plt.show()
```
